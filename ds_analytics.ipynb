```python
import json
from typing import List, Tuple
P = './data/augmented_dataset.json'

with open(P, "r") as f:
    augmented_ds : List[List[List[Tuple[List[List[int]], List[List[int]]]]]] = json.load(f)
```

```python
import numpy as np
```

```python
augmented_ds_np = [
    [
        [
            (np.array(t), np.array(a)) for t, a in example
        ] for example in puzzle
    ] for puzzle in augmented_ds
]
```

```python
del augmented_ds
```

```python
# histogram of width, height and area of task matrix
import matplotlib.pyplot as plt

widths = [aug[0].shape[1] for puzzle in augmented_ds_np for example in puzzle for aug in example]
heights = [aug[0].shape[0] for puzzle in augmented_ds_np for example in puzzle for aug in example]
areas = [aug[0].shape[0] * aug[0].shape[1] for puzzle in augmented_ds_np for example in puzzle for aug in example]


# stats about all of them
max_w, min_w, mean_w, median_w = np.max(widths), np.min(widths), np.mean(widths), np.median(widths)
max_h, min_h, mean_h, median_h = np.max(heights), np.min(heights), np.mean(heights), np.median(heights)
max_a, min_a, mean_a, median_a = np.max(areas), np.min(areas), np.mean(areas), np.median(areas)
print(f'{max_w =}, {min_w =}, {mean_w =}, {median_w = }')
print(f'{max_h =}, {min_h =}, {mean_h =}, {median_h = }')
print(f'{max_a =}, {min_a =}, {mean_a =}, {median_a = }')



plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(widths, bins=20, color='blue', alpha=0.7)
plt.title('Width Histogram')
plt.xlabel('Width')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(heights, bins=20, color='green', alpha=0.7)
plt.title('Height Histogram')
plt.xlabel('Height')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(areas, bins=20, color='red', alpha=0.7)
plt.title('Area Histogram')
plt.xlabel('Area')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
